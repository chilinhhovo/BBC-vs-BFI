```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Fetch the URL and parse the page
url = "https://bfi.org.uk/sight-and-sound/greatest-films-all-time"
response = requests.get(url)
soup_doc = BeautifulSoup(response.content, "html.parser")

# Find all articles
articles = soup_doc.find_all('article', class_="PreviewCard__Article-sc-7y1ad6-0 jREFca")

# Initialize lists for films, directors, and years
films = []
directors = []
years = []

# Extract film titles, directors, and years
for article in articles:  # Iterate through individual articles
    # Extract film title
    film_title = article.find('h1').get_text(strip=True) if article.find('h1') else None

    # Extract all <p> elements with the specified class
    details = article.find_all('p', class_="ResultsPage__P-sc-of10co-2 dgCiym")
    
    # Extract year and director based on their positions
    year = details[0].get_text(strip=True) if len(details) > 0 else None
    director = details[1].get_text(strip=True).replace("Directed by ", "") if len(details) > 1 else None

    if film_title:
        films.append(film_title)
        directors.append(director)
        years.append(year)

# Create the BFI dictionary
BFI = {
    "Films": films,
    "Directors": directors,
    "Year": years
}

# Convert to DataFrame
df = pd.DataFrame(BFI)

# Display the DataFrame
print(df)
```

```python
import pandas as pd 
df = pd.DataFrame(BFI)
print(df.to_string(index=False))
```

```python
import re 
df['Year1'] = df['Year'].apply(lambda x: re.search(r'(\d{4})', x).group(1) if re.search(r'(\d{4})', x) else None)
df['Country'] = df['Year'].apply(lambda x: re.search(r'(\D.*)', x).group(1).strip() if re.search(r'(\D.*)', x) else None)
df['Director'] = df['Directors']
df = df.drop(columns=['Directors'])
df.head()
```

```python
df.to_csv("BFI.csv",index=False)
print("converted and saved!")
```
